```python
import time
from pyfirmata import Arduino, util

a = Arduino('COM5')
lcd = a.get_pin('d:5:o')
b_entry = 'd:6:i'
b_exit = 'd:7:i'
redpin = 'd:9:p'
greenpin = 'd:10:p'
bluepin = 'd:11:p'
s = a.get_pin('d:8:s')

flag1 = 0
flag2 = 0
Total = 13
Space = Total

while True:
    pin_value_1 = a.digital[b_entry].read()
    pin_value_2 = a.digital[b_exit].read()

    if pin_value_1 == 1 and flag1 == 0:
        if Space > 0:
            flag1 = 1
            s.write(0)
            Space += 1
            a.digital[redpin].write(0)
            a.digital[greenpin].write(0)
            a.digital[bluepin].write(1)
        else:
            print('Plz come later')
            print('Available: 0')
            time.sleep(1)
            lcd.write('                ')
    if pin_value_2 == 1 and flag2 == 0:
        flag2 = 1
        s.write(0)
        Space -= 1
        a.digital[redpin].write(0)
        a.digital[greenpin].write(0)
        a.digital[bluepin].write(1)

    if flag1 == 1:
        time.sleep(1)
        s.write(1)
        a.digital[redpin].write(0)
        a.digital[greenpin].write(1)
        a.digital[bluepin].write(0)
        flag1 = 0

    lcd.write('Welcome!!!')
    lcd.write('Available: ')
    lcd.write(str(Space))
    lcd.write(' ')

def RGB_color(red_light_value, green_light_value, blue_light_value):
    a.digital[redpin].write(red_light_value)
    a.digital[greenpin].write(green_light_value)
    a.digital[bluepin].write(blue_light_value)
```
